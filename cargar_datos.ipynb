## Tabla contrato

```python
df_contrato=df[['department','businesstravel','joblevel','jobrole','stockoptionlevel','trainingtimeslastyear','remotework']]
```

```python
contrato=[tuple(i) for i in df_contrato.values]
```

```python
for cont in tqdm(contrato):
    cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                                host='localhost', port='3306',
                                database='base_datos_empleados')


    mycursor = cnx.cursor()
    query = "INSERT INTO datos_contrato (department ,business_travel, job_level, job_role, stock_option_level, training_times_last_year, remote_work) VALUES (%s, %s, %s, %s, %s, %s, %s)"
    val = (cont[0],cont[1],cont[2],cont[3],cont[4],cont[5],cont[6],)
    try: 

        mycursor.execute(query, val)
        cnx.commit()
        print(mycursor.rowcount, "registro insertado.")

    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```
